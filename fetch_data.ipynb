```python
#CONNECT TO THE SQL SERVER
import mysql.connector
from mysql.connector import Error,connect
def create_connection():
  connection=connect(
    host="localhost",
    user="root",
    password="root",
    database="logistics"
  )
  try:
   if connection.is_connected():
    print("Successfully connected to the database")
    return connection
  except Error as err:
    print(f"Error: {err}")
    return None



create_connection()
```

```python
#WRITE THE CSV FILES TO THE SQL TABLES IN MYSQL SERVER
import pandas as pd
from sqlalchemy import create_engine

#  MySQL connection string
engine = create_engine("mysql+mysqlconnector://root:root@localhost/logistics")

# Write the DataFrame cost_df to SQL
costs_df.to_sql(
    name="costs",        # table name
    con=engine,          # connection/engine
    schema="logistics",  # schema (optional, depends on DB)
    if_exists="replace", 
    index=False          # don’t write DataFrame index as a column
)
```

```python

# Write the DataFrame courier_staff_df to SQL
courier_staff_df.to_sql(
    name="courier_staff",        # table name
    con=engine,          # connection/engine
      
    if_exists="replace", 
    index=False          # don’t write DataFrame index as a column
)
```
